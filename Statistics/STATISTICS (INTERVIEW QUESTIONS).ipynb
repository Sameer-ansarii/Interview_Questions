### Q) What is cosine similarity and their uses in ML?

* Cosine similarity is a measure of similarity between 2 data points in a multi-dimensional space.


* Cosine similarity is the angle between vectors in a vector space.


* Cosine similarity doesn't depend on the magnitude of the vectors, it's only depends on angle between vectors.


* It is calculated by taking the dot product of the vectors and dividing it by the product of their lengths.


* The value of cosine similarity ranges between 0-1. The value closer to 1 indicates the more similarity between vectors and the value closer to 0 indicates the more disimilarities between vectors.


Uses:- Cosine similarity can be used to measure the similarity between two data points, two variables in a dataset, two images, two pieces of text, or two other types of data.


Formula:- Cos(x, y) = x . y / ||x|| * ||y||

where,

x . y = product (dot) of the vectors ‘x’ and ‘y’.


||x|| and ||y|| = length of the two vectors ‘x’ and ‘y’.


||x|| * ||y|| = cross product of the two vectors ‘x’ and ‘y’.


* The cosine similarity between two vectors is measured in ‘θ’.


* If θ = 0°, the ‘x’ and ‘y’ vectors overlap, means they are similar.


* If θ = 90°, the ‘x’ and ‘y’ vectors are dissimilar.

### Q) What are the different types of sampling techniques?

Sampling techniques are used to collect samples from a population. The goal of sampling is to create samples that represents the entire population. 

1) Simple Random Sampling :- In simple random sampling we collect samples randomly, it means every person has an equal chance to select as a sample.


2) Stratified sampling: In stratified sampling we first categorize the entire population than collect the samples from the selected group.


3) Systematic sampling: In systematic sampling every kth member is selected as a sample. For example the entire population is 1000 and you have to collect only 100 samples than you have to collect samples from every 10th person.


4) Convenience Sampling :- In convience sampling samples are collected based on convenience and accessibility, it means we collect samples from that individuals who are easily available.


5) Cluster Sampling: In cluster sampling we first make clusters of entire population than collect samples from random clusters.

### Q) What is measures of dispersion ?

1) Range:- Range describes the minimum and maximum value of the data. The range is calculated by taking the difference between the maximum and minimum values of a data


2) Variance:- Variance is a measure to check the spread of data and their distribution. It is calculated by taking the average of the squared differences between each data point and the mean value. A high variance indicates that the data is spread out over a wide range, while a low variance indicates that the data is concentrated around the mean. The value of variance is in square form.


3) Standard Deviation: Standard deviation is a measure to check the spread and distribution of data around it's mean. Standard deviation is the square root of variance. Standard deviation is use to normalize the value of variance. 

* Standard deviation has a concept of Z-Score which is also known as empirical rule or empirical formula use to understand the distribution of data based on standard deviation, which says that our data is present in between -3 to +3 standard deviation range where 68% data present between -1 to +1 standard deviation range, 95% data present between -2 to +2 standard deviation range, 99.7% data present between -3 to +3 standard deviation range and remaining 0.3% data present outside the -3 to +3 standard deviation range and consider as outliers.

### Q) What is Quartile?

3) Quartile 3 (Q3):- (51-75) percentile range, it indicates that 75% of data is below this point.


4) Quartile 4 (Q4):- (76-100) percentile range, it indicates that 100% of data is below this point.


USe:- Quartile are used to identify the distribution and variability of the data within a specific range.

### Q) What is right or positive skewness?

* Right skewness, also known as positive skewness. It occurs when majority of the data points are the lower values.


* In case of right skewness the mean is greater than the median.


* Another reason for right skewness is the presence of outliers, which are higher values that pull the mean towards the right side.


* For example, in an income variable, if the majority of people have lower incomes but there are a few individuals with significantly higher incomes, it can lead to right skewness.

### Q) What is left or negative skewness?

* Left skewness, also known as negative skewness. It occurs when majority of the data points are higher values.


* In case of left skewness the mean is smaller than the median.


* Another reason for left skewness is the presence of outliers, which are lower values that pulls the means towards the left side


* Eg:- We have variable of test scores where the majority score are higher values, and very few students have low scores, it can lead to left skewness.

### Q) What are the types of rule in proabability?

We have 2 types of rule in probability:
    
    
1) Additive Rule (or rule): In additive rule we find probability of event A and event B by adding them. We have 2 types of additional rule.


a) Addition Rule for Mutually Exclusive Events: This rule is apply to find the probability of Mutually Exclusive Events. Mutually Exclusive Events means either event A happening or event B happening, both are not happening together.


b) Addition Rule for Non Mutually Exclusive Events: This rule is apply to find the probability of Non Mutually Exclusive Events. 
Non-Mutually Exclusive Events means both event A and event B happens together.


2) Multiplicative Rule (and Rule) / Joint Probability: In Multiplicative rule we find probability of 2 or more events happening together at same time by Multiplying their probabilities together.

    Mathematically, for mutually exclusive events:
    
    P(A or B) = P(A) + P(B)

    Question: For example, consider rolling a six-sided die. The probability of getting a 2 or a 4 is the sum of the individual probabilities:
    
    P(2 or 4) = P(2) + P(4) 
              = 1/6 + 1/6 
              = 1/3
              = 0.33

    Mathematically, for non-mutually exclusive events:

    P(A or B) = P(A) + P(B) - P(A and B)

    Question: For example, consider drawing a card from a standard deck of 52 cards. The probability of drawing either a heart or a king is the sum of their individual probabilities minus the probability of drawing the king of hearts:
    
    P(heart or king) = P(heart) + P(king) - P(king of hearts) 
                         = 13/52 + 4/52 - 1/52 
                         = 16/52 
                         = 4/13
                         = 0.30

    Mathematically, for Multiplicative Rule (and Rule):

    Suppose we have a bag containing 4 red balls and 6 blue balls. We randomly draw two balls from the bag without replacement. We want to find the probability of selecting a red ball on the first draw and a blue ball on the second draw.

    To solve this problem, we can apply the Multiplicative Rule as follows:

    Probability of drawing a red ball on the first draw:
    P(Red on 1st draw) = Number of red balls / Total number of balls
                       = 4 / 10
                       = 2 / 5

    Probability of drawing a blue ball on the second draw, given that a red ball was drawn on the first draw:
    P(Blue on 2nd draw | Red on 1st draw) = Number of blue balls / Remaining number of balls after the first draw
                                          = 6 / 9
                                          = 2 / 3

    Applying the Multiplicative Rule (and Rule):
    P(Red on 1st draw and Blue on 2nd draw) = P(Red on 1st draw) * P(Blue on 2nd draw | Red on 1st draw)
                                            = (2 / 5) * (2 / 3)
                                            = 4 / 15

    Therefore, the probability of selecting a red ball on the first draw and a blue ball on the second draw is 4/15.
