Q) What is PCA?


* PCA (Principal Component Analysis) is a statistical technique used for reducing the dimensionality of a large dataset while retaining as much information as possible.


* In other words, it is a method to use to convert high-dimensional data into a smaller number of dimensions, called principal components, which capture the most important information in the data. 

Q) Why we use PCA? / What is the main objective of PCA?


1) Dimensionality reduction: PCA is used to reduce the number of variables in a dataset while retaining the most important information.


2) Data visualization: It is hard to visualize data which is above 3D. By reducing the dimensions of a dataset, PCA can help to visualize the data in a lower-dimensional space. It is impossible to visualize data above 3D. 


3) Feature extraction: In high dimension datasets there are variables which are irrelevant and it is difficult to check the variable importance for each variable and select a subset of features, PCA reduces the dimension of dataset while retaining all the variance, this can be helpful to choose a subset of principal components which carries most of the variance.


4) Addressing Colinearity and multicollinearity: Principal Components doesn't conatin colinearity and multicollinearity.

Q) Explain the steps involved in performing PCA.

The steps involved in performing PCA are:-


1) Standardize the data :- Use the Standard Scaler and scale the data to ensure that all features are on the same range.


2) Compute the covariance matrix: The next step is to compute the covariance matrix of the standardized data. Convariance matrix tells the variance and covariance between each pair of variables.


3) Compute the eigenvalues: The third step is to compute the eigenvalues with the help of covariance matrix. Eighen values are equivalent to the number of featutres in the dataset.


4) Compute the eighenvectors: Compute the eighenvectors with the help of eighen values and covariance matrix. 


5) Choose the optimal number of principal components: Choose the optimal number of principal components according the amount of variation they explained.


6) Transform the data: The final step is to transform the data using the chosen number of principal components.


Note: Eigenvectors are known as Principal Components and Eigenvalues are amount of variance that each eighenvector explains.

Q) What is covariance-matrix?


* Covariance indicates the directional relationship between variables. Whether it is positive or inverse or we can say how much two variables change together.


* In PCA covariance matrix is computed by using the scaled values of original features.


* The covariance matrix is a square matrix that shows the covariance between each pair of variables in a dataset.


* The diagonal values of the covariance matrix represent the variance between each pair of variables.


* The off-diagonal values represent the covariance between each pair the variables.

Q) How to select the optimal number of Principal Components?

1) Scree Plot: It is a visualization method by which find optimal number of Principal Components. We plot a scree plot where x-axis defines the number of principal compents and y-axis define the cumulative Explained Variance(CEV) also known as explained variance ratio(EVR). We select that subset of Principal Components which carries maximum variance of the data.


2) Select a threshold for cumulative Explained Variance(CEV). i.e- If we select the threshold as 0.90, then it will return only that subset of Principal Components which explains minimum 90% variance of data.
