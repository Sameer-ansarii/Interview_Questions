### Q) What is linear regression, and when would you use it?



* Linear Regression is a statistical method and supervised Machine Learning algorithm in which the model finds the best fit  which models the linear relationship between the dependent and independent variables.


When to use linear regression :-


* When the data have linear or sort of linear relationship.


* When yo have to predict numerical value which is continuous in nature. Eg:- Price of something like flight fares, product costs etc, or Calculation resale value of something like mobiles,cars and bikes, etc.


### Q) Formul of Linear Regression?

     y = mx+c for simple regression and y = (M1*X1)+(M2*X2)+(m3*x3)....(Mn*Xn) + c for multiple regression
    
     y = dependent variable
     m = coefficient/sclope
     x = independent variable
     c = intercept/constant


### Q) How linear regression finds the value of slope and intercept?

* Linear regression used ordinary least square method (OLS) to find the value of slope and intercept.


     * Formula for Slope:-
        
     m = (∑(y - y_mean)(x - x_mean)) / (∑(x - x_mean)^2)
    
     where:

     m = slope
     y = dependent variable
     x = independent variable
     y_mean = of the y values
     x_mean = mean of the x values
    
    
    * Fromula of Intercept
    
     b = y_mean - m*x_mean
     b = intercept
     m = slope

### Q) What is slope / coefficient / gradient in linear regression?


* Slope measures the rate of change between 2 variables or we can say how 2 variables change together is known as slope.


* Value of slope comes in positive, negative or 0. If the value of slope is 2, it means when we increase the value of x by 1 unit then value of y will increase by 2 units. When the value of slope is -2 it means when we increase the value of x by 1 unit the value of y will decrease by 2 units.


* For each variable we have different slope value.

### Q) What is residual and what is residual error? 


* The difference between the actual and predicted values of the dependent variable is called residual. 


* The sum of squared residuals / error is called residual error. 

### Q) What is loss function and cost function?

* Loss Function is known as error function, it is calculated by taking the squared difference between actual value and predicted value. When we calculate the error of a single observation it is called loss function.


* Cost function is the overall error of a dataset. We calculate cost function by averaging the squared difference between actual value and predicted value for all observations, it is called as cost function.

### Q) How to find best fit line, what is best fit line and use of best fit line?


* Linear regression models uses ordinary least square method (OLS) to find the best fit line. Model creates multiple straight lines and calculate sum of square error for which line the sum of square error is minimum, model select that line as a best fit line.


* Motive of this best fit line is to reduce the cost function.

### Q) What is the objective of linear regression, and how is it achieved?


* The objective of linear regression is to build the relationship between the dependent variable and independent variables by creating a best fit line which minimizes the sum of squared error.


### Q) What is the difference between simple and multiple linear regression?

* In multiple regression we have multiple independent variables and according to all those variable we have to predict the vlue of dependent variable.

### Q) What is Mean Squared Error(MSE) and how to find it and what is it use?


* MSE stands for Mean Squared Error, it measure the average squared difference between actual value and predicted value.


How to find MSE:-

* Calcualate the squared difference between actual value and predicted value for all observations and calculate it's average.

Use:-


* It is use to find the best fit line and it is also use for model evaulation


### Q) What is Root Mean Squared Error(RMSE)?

### Q) What is MAE?

### Q) What is R-Squared and it formula?


* R-squared is also known as goodness of fit and coefficient of determination because it measures that how well our best fit line captures the variability of data or we can say R-Squared measures the proporation of variance in the dependent variable that can be explained by one or more independent variables.


* The value of R-squared is ranges between 0-1, the value closer to 1 indicates that our best fit line captures the variablity of data too well.

    Sum of Squared residual = squared difference between actual value and predicted value
    
    Sum of Squared Total = Squared difference between mean value and actual value


### Q) What is adjusted R-Squared and it's formula?

* Ajusted R-Square is a improved version of R-Squared and it is more accurate goodness of fit of a linear regression model and prevents model from overfitting.


* Problem with R-Square is that when we increase the number of variables whether they contribute in model or not, R-squared value is increasing and this will lead to overfitting.


* Ajusted R-Sqaure value is decreasing if we put any variable that do not contribute in model.If R-squared value is high and adjusted R-suared value is very low in that case model leads to overfitting. So, by comparing R-Squared value with the Adjusted R-Sqaured value we prevent model from overfitting.


* Value of Adjusted R-Square is ranges between 0-1 and always lower or equals to R-Squared value.


    Formula:
    
    adjusted R-squared = 1 - (1 - R-squared) * (n - 1) / (n - p - 1)
    
    n = number of observations
    
    
    p = number of independent variables

### Q) What is regularization?


Regularization are the techniques which are used to prevent model from overfitting and multicollinearity. We have three techniques of regularization in linear regression, such as ridge regression, lasso regression and elastic net regression.

### Q) What is Ridge Regression?


* Ridge Regression is a regularization technique which is also known as L2 Regularization. It is used to reduce overfitting and multicolinearity in a linear regression model.


* Ridge regression adds a pentaly term to the cost function of linear regression, which is lamda * submission of squared slopes by adding this penalty term, regression line will modified and the problem of high variance multicollinearity will reduce.

### Q) What is Lasso Regression?

* Lasso is a regularization technique, also known as L1 regularization. Lasso regression is used to reduce multicolinearity and overfitting in a linear regression model.


* Lasso regression add a penalty term in cost function of linear regression which is lamda * submission of  mod of slopes. By adding this penalty term regression line will modified and the problem of overfitting and multicolinearity will reduce.

### Q) What is elastic net regression?


* It is a regularization technique used to reduce the multicolinearity and overfitting in a linear regression model.


* Elastic net regression adds the penalty term of ridge regression and lasso regression to the cost function of linear regression, which is cost function of linear regression + lamda * submission of squared slopes + lamda into submission of mod of slopes. By adding these penality term regression line will modified and the problem of high variance and multicolinearity will reduce.

### Q) What are the assumptions of linear regression?

1) Linearity: The relationship between all independent variables and dependent variables are linear.

    
2) Normality: Residuals should be normally distributed.


3) Independence: All observations are independent to each other, it means one observation do not impact the result of another observation.


4) Homoscedasticity: Homo means same and scedasticity means scatter. It means variance of residuals and predicted values should be almost same.


5) No multicollinearity: The independent variables should not be highly correlated with each other.

## Q) Advantages and Disadvantages of Linear regresion model?


Advantages:-


* Simple and easy to understand


* Fast computation


* Linear regression performs exceptionally well for linearly separable data.


Disadvantages :- 


* Linear regression assumes that the relationship between the dependent and independent variables is linear, which may not always be the case.


* Linear regression is sensitive to outliers in the data, which can have a large impact on the model's performance.
