### Q) What is MYSQL and it's use?

`MySQL` is a open-source and free `RDMS (relational database management system)` used to `store, manage and manipulate databases`.

### Q) Difference between SQL and MYSQL.

`SQL (Structured Query Language)` is a programming language used to manage and manipulate relational databases.

`MySQL`, on the other hand, is a specific `relational database management system (RDMS)` that uses SQL as its language for `interacting with the database`. `MySQL` is an `open-source` and `free RDMS` that implements the SQL language and provides the software and tools to `store, manage, and retrieve data.`

### Q) Difference between SQL, My SQL and No SQL?

`MySQL` is a relational database management system that uses `SQL`. Whereas `SQL` is primarily used to query and operate database systems, `MySQL` enables you to store, handle, delete, and modify data in an organized way.

As for `NoSQL`, it is a non-relational database that does not use `SQL`. `NoSQL` databases use a simpler data structure, meaning they are quicker, more responsive, and more scalable than traditional relational databases.

### Q)  Difference between RDMS and Non-RDMS?

**`RDMS`**

**`Data Model`**: RDBMS uses a structured data model based on tables, rows, and columns, where data is organized into predefined schemas and tables with relationships defined through keys.


**`Query Language`**: RDBMS typically uses SQL (Structured Query Language) to manipulate and retrieve data from the database.


**`Examples`**: MySQL, Oracle Database, PostgreSQL, Microsoft SQL Server.


**`Non-Relational Database Management System (NoSQL or Non-SQL DBMS)`**

**`Data Model`**: NoSQL databases use various data models like key-value, document, columnar, and graph. These models are more flexible and dynamic, allowing unstructured and semi-structured data to be stored.


**`Query Language`**: NoSQL databases use different query languages depending on the specific database type, such as MongoDB's query language for document databases.


**`Examples`**: MongoDB, Cassandra, Redis, Neo4j.

### Q) What are Constraints?

 Constraints are used to specify rules for data in a table


 This ensures the accuracy and reliability of the data in the table


 Constraints can be specified when the table is created with the CREATE TABLE statement, or after the table is created with the ALTER TABLE statement

 **`Syntax`**

CREATE TABLE table_name(

column1 datatype **`constraint`**,

column2 datatype **`constraint`**,

**`Constraints`**

Commonly used constraints in SQL:

* `NOT NULL` - Ensures that a column cannot have a NULL value


* `NIQUE` - Ensures that all values in a column are different


* `PRIMARY KEY` - A combination of a NOT NULL and UNIQUE


* `FOREIGN KEY` - Prevents actions that would destroy links between tables (used to link multiple tables together)


* `CHECK` -Ensures that the values in a column satisfies a specific condition


* `DEFAULT` - Sets a default value for a column if no value is specified


* `CREATE INDEX` - Used to create and retrieve data from thedaveryquickly

### Q) What is the use of `WHERE` Clause?

`WHERE` Clause is used to filter records. It is used to extract only those records that fulfil a specified condition. 

### Q) What is the use of `LIMIT` Clause?

The `LIMIT` clause is used to specify the number of records to return.

The `LIMIT` clause is useful on large tables with thousands of records. Returning a large number of records can impact performance.

### Q) What is the use of `ORDER BY` Clause?
`Order By` keyword is used to sort the values of a column in ascending order.


The `ORDER BY` keyword is used to sort the result-set in ascending or descending order.


The `ORDER BY` keyword sorts the records in ascending order by default. To sort the records in descending order, use the `DESC` keyword.

### Q) What is the use of `GROUP BY` Statement? 

The `GROUP BY` statement groups rows that have the same values into summary rows.


It is often used with aggregate functions `COUNT()`, `MAX()`, `MIN()`, `SUM()`, `AVG()` to group the result-set by one or more columns.


**NOTE:-**

* All (*) is not used with `GROUP BY`.


* `GROUP BY` Statement is used with aggregate functions.


* `HAVING` Clause is used with `GROUP BY`, if you use `WHERE` Clause with `GROUP BY` it will show error.


* The column we use with `select` statement is also comes with `group by`. EG : **Select `Mode`, SUM(Amount) From Payment GROUP BY `Mode`;**


### Q) What is the use of `HAVING` Clause?

The `HAVING` clause is used to apply a filter on the result of GROUP BY based on the specified condition.


The `WHERE` clause places conditions on the selected columns, whereas the `HAVING` clause places conditions on groups created by the`GROUPBY`clause.


The `HAVING` clause was added to SQL because the `WHERE` keyword cannot be used with `group by` & aggregate functions.

### Q) What is `JOIN` and their types?

Join means combine something.


A `JOIN` clause is used to combine data from two or more tables, based on a related column between them.


**TYPES OF JOINS**


1) **`INNER JOIN`** : Returns records that have matching values in both tables


2) **`LEFT JOIN`** / **`LEFT OUTER JOIN`** : Returns all records from the left table, and the matched records from the right table. If matched records are not found it will show NULL on the right table.


3) **`RIGHT JOIN`** / **`RIGHT OUTER JOIN`** : Returns all records from the right table, and the matched records from the left table.  If matched records are not found it will show NULL on the left table.


4) **`OUTER JOIN`** : Returns all records from both tables. If there is no matches values are present it will shown NULL over their.


5) **`SELF JOIN`** 


* A self join is a regular join in which a table is joined to itself.


* Self Join are powerful for comparing values in a column of rows with the same table.

### Q) What is the use of `UNION` Operator?


The `UNION` operator is used to combine the result-set of two or more `SELECT` statements exclusing duplicates.

### Q) What is the ue of `CASE` Statement?

The `CASE` statement goes through conditions and returns a value when the first condition is met (like an if-then-else statement). So, once a condition is true, it will stop reading and return the result. If no conditions are true, it returns the value in the ELSE clause.

If there is no `ELSE` part and no conditions are true, it returns NULL.

**`CASE`**
    
   **`WHEN` condition1 `THEN` result1**
    
   **`WHEN` condition2 `THEN` result2**
   
   **`WHEN` conditionN `THEN` resultN**
    
   **`ELSE` result**

**`END`;**
