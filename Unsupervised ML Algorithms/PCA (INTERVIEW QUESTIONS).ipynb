Q) What is PCA?


* PCA (Principal Component Analysis) is a statistical technique used for reducing the dimensionality of a large dataset while retaining as much information as possible.


* In other words, it is a method to use to convert high-dimensional data into a smaller number of dimensions, called principal components, which capture the most important information in the data. 

Q) Why we use PCA? / What is the main objective of PCA?


1) Dimensionality reduction: PCA is used to reduce the number of variables in a dataset while retaining the most important information.


2) Data visualization: It is hard to visualize data which is above 3D. By reducing the dimensions of a dataset, PCA can help to visualize the data in a lower-dimensional space. It is impossible to visualize data above 3D. 


3) Feature extraction: In high dimension datasets there are variables which are irrelevant and it is difficult to check the variable importance for each variable and select a subset of features, PCA reduces the dimension of dataset while retaining all the variance, this can be helpful to choose a subset of principal components which carries most of the variance.


4) Addressing Colinearity and multicollinearity: Principal Components doesn't conatin colinearity and multicollinearity.

Q) Explain the steps involved in performing PCA.

The steps involved in performing PCA are:-


1) Standardize the data :- Use the Standard Scaler and scale the data to ensure that all features are on the same range.


2) Compute the covariance matrix: The next step is to compute the covariance matrix of the standardized data. Convariance matrix tells the variance and covariance between each pair of variables.


3) Compute the eigenvalues: The third step is to compute the eigenvalues with the help of covariance matrix. Eighen values are equivalent to the number of featutres in the dataset.


4) Compute the eighenvectors: Compute the eighenvectors with the help of eighen values and covariance matrix. 


5) Choose the optimal number of principal components: Choose the optimal number of principal components according the amount of variation they explained.


6) Transform the data: The final step is to transform the data using the chosen number of principal components.


Note: Eigenvectors are known as Principal Components and Eigenvalues are amount of variance that each eighenvector explains.

Q) What is covariance-matrix?


* Covariance indicates the directional relationship between variables. Whether it is positive or inverse or we can say how much two variables change together.


* In PCA covariance matrix is computed by using the scaled values of original features.


* The covariance matrix is a square matrix that shows the covariance between each pair of variables in a dataset.


* The diagonal values of the covariance matrix represent the variance between each pair of variables.


* The off-diagonal values represent the covariance between each pair the variables.

Q) How to select the optimal number of Principal Components?

1) Scree Plot: It is a visualization method by which find optimal number of Principal Components. We plot a scree plot where x-axis defines the number of principal compents and y-axis define the cumulative Explained Variance(CEV) also known as explained variance ratio(EVR). We select that subset of Principal Components which carries maximum variance of the data.


2) Select a threshold for cumulative Explained Variance(CEV). i.e- If we select the threshold as 0.90, then it will return only that subset of Principal Components which explains minimum 90% variance of data.

Q) What are Eighenvectors and Eighen Values? / Can you explain the concept of eigenvalues and eigenvectors in PCA?

* Eigenvalues and eigenvectors are concepts of linear algebra that are used in PCA to transform high-dimensional data into a lower-dimensional space.


* Eigenvalues represent the amount of variance or information contained in each eigenvector. They are scalar quantities that indicate the magnitude or amount of variance captured by each eigenvector.


* Eigenvectors represent the directions of variance in the multi-dimensional space. They are vector quantities that indicate both the magnitude and direction of variance.


* Principal Components are also known as Eighenvector because they both represents the direction of variance and Eighenvalue represents the amount of variance captured by each principal component. 
 

* The first principal component is the eigenvector with the largest eigenvalue, and it represents the direction of maximum variance in the data. The second principal component is the eigenvector with the second-largest eigenvalue, and it represents the direction of maximum variance in the data orthogonal(perpendicular) to the first principal component. This process is repeated until all principal components are found.

Note:-

* Magnitude indicates the distance of data point from origin and direction indicates where a data point lies in a multi-dimensional space.

* Orthogonal(Perpendicular) means 90 degree angle between 2 lines.

Q) What is the interpretation of principal components? / Explain Principal Components.


* Principal components are a set of new variables that are equivalent to the number of existing variables, derived from the original features of a dataset.


* They are created in such a way that the first principal component captures the maximum variance in the data, second principal component captures the second maximum variance in the data and so on.


* Principal components are orthogonal(perpendicular) to each other, which means they do not contains colinearity and multi-colinearity.


* Principal components are useful because they allow us to reduce the dimensionality of the data, while retaining most of the important information.

Q) What will happen if we keep increasing the number of principal components?

* As we select more number of principal components, we capture more information from the data, which helps in lossing less informaton.

Q) What is Curse of Dimensionality?


* Curse of Dimensionality is a problem in machine learning which says that there is a optimum limit of features if we cross this optimum limit and put less or more features to build a model, than will not give better accuracy and problem of overfitting and underfitting may arise.


* Usually the problem of Curse of Dimensionality arises while working on a large dataset. To prevent from this problem we use many features selections techniques like RFE, VIF, Correlation Matrix and some Feature Extraction Techniques like PCA.

Q) What is the difference between covariance matrix and correlation matrix in PCA?


* Covariance Matrix: A covariance matrix is a square matrix that shows the variance and covariance between each pair of variable . Diagonally it shows the covariance between each pair of variables and off-diagonally it show the variance. Covariance measures the directional relationship between each pair of variables whether it is positive or inverse. The value of covariance ranges between -infinity to +infinity.


* Correlation Matrix: A correlation matrix is also a square matrix that shows the correlation between each pair of features in a dataset. Correlation measures the strength of directional relationship between variables whether it is positive or inverse. The value of correlation ranges between -0 to +1. 


Q) Can PCA be used for feature selection?

* Yes, PCA can be used as feature selection. We can visualize the Principal Components against the cumulative of Explained Variance Ratio(EVR), which helps in to select a subset of Principal Components which explains maximum variance of data.

Q) What are the similarities between PCA and linear regression?

1) In linear regression R² score indicates the amount of variation explained by all independent variables and in PCA each Principal components explained some amount of variance which is also R² of individual PCA.


2) Linear Regression starts by finding the best fit line and PCA also starts by finding the best fit line, best fit line indicates the direction of maximum variance, and then roatate the best fit line and create new coordinates, which helps in to make other principal components which are orthogonal(perpendicular) to each other.

Q) Advantages of PCA.

1) Helps to reduce the dimension of data while retaining as much information as possible.


2) Help in visualizing multi-dimensional data by converting them into 2 or 3 dimensional data.


3) Removes the colinearity and multi-colinearity among independent variables, this result into reduce overfitting problem.


4) It convert high dimensional data into low dimensional data, that's the second reason to reduce overfitting.


5) We use PCA as a feature selection technique, as we select a subset of principal components which captured maximum variance of data.


6) It handles the noise of data.


7) Model takes less in time in building due to less number of variables.

Q) What are some of the limitations of PCA?


1) Required Scaling: Data needs to be scaled before performing PCA. Otherwise, the result may be biased towards the variables with larger ranges. PCA required the same range of variables.


2) Information loss: By selecting a subset of Principal Components, we loss some amount of information of data.


3) Difficulity in interpretation of variables: It is difficcult to understand the Principal Components rather than understanding the existing variables.
